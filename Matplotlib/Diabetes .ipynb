```python
import pandas as pd

import numpy

import tensorflow
```

```python
db=pd.read_csv("pima-indians-diabetes.data.csv",header=None)
```

```python
db.head(5)
```

```python
y=db[8]
```

```python
x=db.iloc[:,0:8]
```

```python
x
```

```python
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test = tts(x,y,test_size=0.20)
```

```python
from keras.models import Sequential
```

```python
model=Sequential()
```

```python
from keras.layers import Dense
```

```python
model.add(Dense(units=4,kernel_initializer="Zeros",bias_initializer="Zeros", activation="relu",input_dim=8))
```

```python
model.add(Dense(units=4,kernel_initializer="Zeros",bias_initializer="Zeros", activation="relu"))
```

```python
model.add(Dense(units=1,kernel_initializer="Zeros",bias_initializer="Zeros", activation="sigmoid"))
```

```python
from keras.optimizers import Adam 
model.compile(optimizer=Adam(learning_rate=0.00001),loss="binary_crossentropy", metrics="accuracy" )
```

```python
history=model.fit(x_train,y_train, epochs=100)
```

```python
from matplotlib import pyplot as plt
plt.plot(history.history['loss'])
```
